## Doing it all at once now

```python
driver.get("https://www.consumerstartups.com/publish/posts")
```

```python
posts = driver.find_elements(By.CLASS_NAME,"frontend-publish-post_management-shared-postRow-module__postRowLink--pR6Fb")
```

```python
posts = [(post.text.split("\n")[0],post.get_attribute("href")) for post in posts]
```

```python
all_stats = []
post_count = 1
for post in posts:
    driver.get(post[1])
    time.sleep(5)
    stats = driver.find_elements(By.CLASS_NAME, "frontend-pencraft-Box-module__flex-gap-64--dzjmF")
    for stat in stats:
        item = [post_count,post[0]]
        item.extend(stat.text.split("\n")[1:])
        all_stats.append(item)
    post_count+=1
```

```python
all_stats
```

```python
# download csv
import csv
with open("output4.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(all_stats)
print("CSV file has been created!")
```
